```python
%pip install pandas numpy matplotlib scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (us$)')
plt.scatter(df.year, df['per capita income (US$)'],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression() # create an object for linier regression
reg.fit(df[['year']], df['per capita income (US$)']) # training the linear regression model using available data points
#1st argument should be an 2d array in fit method fit(x,y)
```

```python
reg.predict(np.array([[2050]]))
```

for any linier equation we have y=mx+c m is slop or coefficient


```python
reg.coef_#gives value of coeficient
```

```python
reg.intercept_ #gives the intercept
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (us$)')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
plt.plot(df.year, reg.predict(df[["year"]]), color='blue')
```

## generate csv file with list of per capita income prediction

```python
year_df=pd.read_csv("Book1.csv")
print(year_df)
pci = reg.predict(year_df) #predict using year dataframe
year_df['per capita income in $']=pci #insert the predicted value in a column per capita income in $
print(year_df)
```
